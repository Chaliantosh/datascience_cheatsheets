```python
#pull the medals won by athlete 'LEWIS, carl' including the first and most recent olympic wins.
import pandas as pd
oo=pd.read_csv(r"C:\Users\admin\Desktop\Santosh_Learning\Projects in GIThub\Pandas_SummerOlympics_dataanalysis/olympics.csv", skiprows=4)
oo.loc[oo.Athlete == 'LEWIS, Carl'].groupby('Athlete').agg({'Edition': ['min','max','count']})
```

```python
"""
Reshaping: Stack and Unstack are very useful when used in conjuntion with groupby()
Stack: Allows inner columns to move to the rows. Returns a DF and series, it does not have missing values since dropna=True
by default. DataFrame becomes taller.
Unstack: does the reverse of the stack. DataFrame becomes wider. since it has missing values, it has a choice of filling 
it missing values.
"""
```

```python
#Athletes winning medals in Beijing Olympics 100m or 200m track event.
mw = oo[(oo.Edition == 2008) & ((oo.Event == '100m') | (oo.Event == '200m'))]
mw
```

```python
g = mw.groupby(['NOC', 'Gender', 'Discipline', 'Event']).size()
g
```

```python
df = g.unstack(['Discipline', 'Event'])
df
```

```python
df.stack() # DateFrame becomes taller
```

```python
df.stack('Event') #same result as above
```

```python
df
```

```python
df.unstack() #DataFrame gets widened
```

```python
df.unstack('Gender') #same result as above
```

```python
"""
DataVizualizations: Heat maps
Using the Olympics dataset, present a summary of the total medals won by participating countries in the 2008 Olympics.
"""
oo
```

```python
lo = oo[oo.Edition == 2008].groupby(['NOC','Medal']).size()
gby = lo.unstack('Medal',fill_value=0)
gsb = gby.sort_values(['Gold', 'Silver', 'Bronze'],ascending=False) #.plot(kind='bar')
gsb_plot = gsb[['Gold', 'Silver', 'Bronze']] #.plot();
sns.heatmap(gsb_plot);
```

```python
fnl = gsb_plot.transpose() 
fnl
```

```python
plt.figure(figsize=(16,5))
sns.heatmap(fnl);
```

```python
#create your own colormaps
#Q17. Plot the five athletes who have won the most gold medals over the history of the Olympics. When there is a tie, 
#     consider the number of silver medals, and then bronze medals.
g = oo.groupby(['Athlete', 'Medal']).size().unstack('Medal',fill_value=0)
h = g.sort_values(['Gold', 'Silver', 'Bronze'],ascending=False)
i = h[['Gold','Silver','Bronze']]
i.head().plot(kind='bar');
```
